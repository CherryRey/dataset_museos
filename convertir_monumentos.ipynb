```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import chain
%matplotlib inline

```

```python
df_bm = pd.read_csv("edificios_monumentos.csv", sep=";", encoding = 'unicode_escape', engine ='python')
df_bm.head()
```

```python
print("Hay", df_bm.shape[0], "entradas")
print("Hay", df_bm.shape[1], "columnas")
print("las columnas son:", list(df_bm))
```

```python
#obtener el nombre de las columnas en una lista
column_name = df_bm.columns.tolist()
list_column = list(column_name)
list_column
```

```python
#Eliminar las columnas que no me interesan
df_bm_drop = df_bm.drop([
    'PK',
    'PDF',
    'CONTENT-URL',
    'NOMBRE-VIA',
    'CLASE-VIAL',
    'TIPO-NUM',
    'NUM',
    'PLANTA',
    'PUERTA',
    'ESCALERAS',
    'ORIENTACION',
    'PROVINCIA',
    'COD-BARRIO',
    'COD-DISTRITO',
    'COORDENADA-X',
    'COORDENADA-Y',
    'TELEFONO',
    'FAX',
    'EMAIL',
    'TIPO',
    'Unnamed: 27'], axis=1)

df_bm_drop
```

```python
#ver el tipo de las variables
display(df_bm_drop.dtypes)
```

```python
#convertir valores NA en 0
df_bm_drop['CODIGO-POSTAL'] = df_bm_drop['CODIGO-POSTAL'].fillna(0)
#quiero convertir a integer el codigo postal
df_bm_drop['CODIGO-POSTAL'] = df_bm_drop['CODIGO-POSTAL'].astype(int)
df_bm_drop.head(5)
```
